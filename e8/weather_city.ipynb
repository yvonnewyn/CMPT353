```python
labelled = pd.read_csv('monthly-data-labelled.csv')
unlabelled = pd.read_csv('monthly-data-unlabelled.csv')


X = labelled.drop(['city', 'year'], axis=1).values
y = labelled['city'].values
X_train, X_valid, y_train, y_valid = train_test_split(X, y)


model = make_pipeline(
    StandardScaler(),
    SVC(kernel='linear', C=0.1)

)

model.fit(X_train, y_train)
print(model.score(X_valid, y_valid))
```

```python
X_unlabelled = unlabelled.drop(['city', 'year'], axis=1).values
y_predicted = model.predict(X_unlabelled)
y_predicted
```

```python
df = pd.DataFrame({'truth': y_valid, 'prediction': model.predict(X_valid)})
print(df[df['truth'] != df['prediction']])
df


```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train)

model = VotingClassifier([
    ('nb', GaussianNB()),
    ('knn', KNeighborsClassifier(5)),
    ('svm', SVC(kernel='linear', C=0.1)),
    ('tree1', DecisionTreeClassifier(max_depth=4)),
    ('tree2', DecisionTreeClassifier(min_samples_leaf=10)),
])

model.fit(X_train, y_train)
print(model.score(X_valid, y_valid))
df = pd.DataFrame({'truth': y_valid, 'prediction': model.predict(X_valid)})
print(df[df['truth'] != df['prediction']])

```
